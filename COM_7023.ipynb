```python
print("=" * 60)
print("TASK 1: LINEAR ALGEBRA - Traffic Flow Equilibrium")
print("=" * 60)


A = np.array([
    [ 2, -1,  1,  0],
    [ 1,  1,  0,  1],
    [-1,  0,  3, -1],
    [ 0,  1,  1,  1]
], dtype=float)

b = np.array([150, 420, 100, 380], dtype=float)

print("\nCoefficient Matrix A:")
print(A)
print("\nConstants Vector b:")
print(b)


det     = np.linalg.det(A)
rank_A  = np.linalg.matrix_rank(A)
rank_Ab = np.linalg.matrix_rank(np.column_stack([A, b]))

print(f"\nDeterminant of A:   {det:.4f}")
print(f"Rank of A:          {rank_A}")
print(f"Rank of [A|b]:      {rank_Ab}")
print(f"Condition Number:   {np.linalg.cond(A):.4f}")
print(f"\nSince det != 0, a unique solution exists meaning the network is stable")


x = np.linalg.solve(A, b)
print(f"\nTraffic flow solution:")
print(f"  x1 (Road 1) = {x[0]:.2f} vehicles per 10 min")
print(f"  x2 (Road 2) = {x[1]:.2f} vehicles per 10 min")
print(f"  x3 (Road 3) = {x[2]:.2f} vehicles per 10 min")
print(f"  x4 (Road 4) = {x[3]:.2f} vehicles per 10 min")


print(f"\nVerification - Ax minus b should be close to zero:")
print(np.dot(A, x) - b)


print(f"\nInverse of A:")
print(np.round(np.linalg.inv(A), 4))


fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle("Task 1: Traffic Flow Equilibrium", fontsize=14, fontweight='bold', color=C1)

bars = axes[0].bar(['Road x1','Road x2','Road x3','Road x4'], x,
                   color=[C1, C2, C3, C4], edgecolor='white', linewidth=1.5)
axes[0].axhline(np.mean(x), color=C2, linestyle='--', linewidth=2, label=f'Mean: {np.mean(x):.1f}')
for bar, val in zip(bars, x):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 3,
                 f'{val:.1f}', ha='center', fontweight='bold', color=C1)
axes[0].set_ylabel("Vehicles per 10 minutes", color=C1)
axes[0].set_title("Equilibrium Traffic Flows", color=C1, fontweight='bold')
axes[0].legend(facecolor='white', edgecolor=C1)
axes[0].tick_params(colors=C1)

im = axes[1].imshow(A, cmap='RdYlBu_r', aspect='auto', vmin=-1, vmax=3)
axes[1].set_xticks(range(4)); axes[1].set_yticks(range(4))
axes[1].set_xticklabels(['x1','x2','x3','x4'], color=C1)
axes[1].set_yticklabels(['Eq.1','Eq.2','Eq.3','Eq.4'], color=C1)
axes[1].set_title("Coefficient Matrix A", color=C1, fontweight='bold')
for i in range(4):
    for j in range(4):
        axes[1].text(j, i, int(A[i,j]), ha='center', va='center',
                     fontweight='bold', fontsize=14, color='white')
plt.colorbar(im, ax=axes[1])
plt.tight_layout()
plt.savefig('task1_traffic_flow.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
print("\n" + "=" * 60)
print("TASK 2: CALCULUS - Velocity and Acceleration")
print("=" * 60)

t = np.linspace(0, 20, 1000)

v = 70 + 12*np.cos(np.pi*t/5) - 4*np.sin(2*np.pi*t/5)

a = -12*(np.pi/5)*np.sin(np.pi*t/5) - 4*(2*np.pi/5)*np.cos(2*np.pi*t/5)

print("\nv(t) = 70 + 12cos(pi*t/5) - 4sin(2*pi*t/5)")
print("a(t) = -(12pi/5)sin(pi*t/5) - (8pi/5)cos(2*pi*t/5)")
print(f"\nVelocity range:     {v.min():.2f} to {v.max():.2f} km/h")
print(f"Acceleration range: {a.min():.4f} to {a.max():.4f} km/h/s")
print(f"In m/s2:            {a.min()/3.6:.4f} to {a.max()/3.6:.4f}")

idx_max = np.argmax(a)
idx_min = np.argmin(a)
print(f"\nPeak acceleration:  {a[idx_max]:.4f} km/h/s at t = {t[idx_max]:.2f}s")
print(f"Peak deceleration:  {a[idx_min]:.4f} km/h/s at t = {t[idx_min]:.2f}s")

fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
fig.suptitle("Task 2: Velocity and Acceleration along Smart Corridor",
             fontsize=14, fontweight='bold', color=C1)

axes[0].plot(t, v, color=C1, linewidth=2.5, label='v(t)')
axes[0].axhline(70, color=C2, linestyle='--', linewidth=1.5, alpha=0.8, label='Mean 70 km/h')
axes[0].fill_between(t, 70, v, where=(v >= 70), alpha=0.15, color=C2, label='Above mean')
axes[0].fill_between(t, 70, v, where=(v < 70),  alpha=0.15, color=C3, label='Below mean')
axes[0].set_ylabel("Velocity (km/h)", color=C1)
axes[0].set_title("v(t) = 70 + 12cos(pi*t/5) - 4sin(2*pi*t/5)", color=C1, fontweight='bold')
axes[0].legend(facecolor='white', edgecolor=C1)
axes[0].tick_params(colors=C1)

axes[1].plot(t, a, color=C2, linewidth=2.5, label='a(t) = dv/dt')
axes[1].axhline(0, color=C1, linewidth=1)
axes[1].fill_between(t, 0, a, where=(a >= 0), alpha=0.2, color=C2, label='Acceleration')
axes[1].fill_between(t, 0, a, where=(a < 0),  alpha=0.2, color=C3, label='Deceleration')
axes[1].annotate(f'Max\n{a[idx_max]:.2f}', xy=(t[idx_max], a[idx_max]),
    xytext=(t[idx_max]+0.7, a[idx_max]+0.6), fontsize=9, color=C2, fontweight='bold',
    arrowprops=dict(arrowstyle='->', color=C2, lw=1.5))
axes[1].annotate(f'Min\n{a[idx_min]:.2f}', xy=(t[idx_min], a[idx_min]),
    xytext=(t[idx_min]+0.7, a[idx_min]-1), fontsize=9, color=C3, fontweight='bold',
    arrowprops=dict(arrowstyle='->', color=C3, lw=1.5))
axes[1].set_ylabel("Acceleration (km/h/s)", color=C1)
axes[1].set_xlabel("Time (seconds)", color=C1)
axes[1].set_title("a(t) = -(12pi/5)sin(pi*t/5) - (8pi/5)cos(2*pi*t/5)", color=C1, fontweight='bold')
axes[1].legend(facecolor='white', edgecolor=C1)
axes[1].tick_params(colors=C1)
plt.tight_layout()
plt.savefig('task2_velocity_acceleration.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
print("\n" + "=" * 60)
print("TASK 3: PROBABILITY - AV Reaction Times")
print("=" * 60)

mu    = 0.85
sigma = 0.25
print(f"\nR ~ N(mu={mu}, sigma={sigma})")

p_rapid   = stats.norm.cdf(0.5, mu, sigma)
p_delayed = 1 - stats.norm.cdf(1.5, mu, sigma)
p_normal  = stats.norm.cdf(1.5, mu, sigma) - stats.norm.cdf(0.5, mu, sigma)

print(f"\nP(R < 0.5s)  rapid reaction:    {p_rapid:.4f}  ({p_rapid*100:.2f}%)")
print(f"P(R > 1.5s)  delayed reaction:  {p_delayed:.4f} ({p_delayed*100:.2f}%)")
print(f"P(0.5 to 1.5s) normal range:    {p_normal:.4f}  ({p_normal*100:.2f}%)")

print(f"\nPercentile values:")
for pct in [5, 10, 25, 50, 75, 90, 95, 99]:
    val = stats.norm.ppf(pct/100, mu, sigma)
    print(f"  {pct:>3}th percentile = {val:.4f}s")

x_range = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y_pdf   = stats.norm.pdf(x_range, mu, sigma)
pvals   = [stats.norm.ppf(p/100, mu, sigma) for p in [5,10,25,50,75,90,95,99]]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle("Task 3: AV Reaction Time Distribution R ~ N(0.85, 0.25^2)",
             fontsize=13, fontweight='bold', color=C1)

axes[0].plot(x_range, y_pdf, color=C1, linewidth=2.5)
axes[0].fill_between(x_range, y_pdf, where=(x_range < 0.5),
                     alpha=0.6,  color=C2, label=f'P(R<0.5s) = {p_rapid*100:.2f}%')
axes[0].fill_between(x_range, y_pdf, where=((x_range >= 0.5) & (x_range <= 1.5)),
                     alpha=0.25, color=C3, label=f'P(0.5-1.5s) = {p_normal*100:.2f}%')
axes[0].fill_between(x_range, y_pdf, where=(x_range > 1.5),
                     alpha=0.6,  color=C5, label=f'P(R>1.5s) = {p_delayed*100:.2f}%')
axes[0].axvline(mu,  color=C1, linestyle='--', linewidth=2, label=f'mu = {mu}s')
axes[0].axvline(0.5, color=C2, linestyle=':', linewidth=1.5)
axes[0].axvline(1.5, color=C5, linestyle=':', linewidth=1.5)
axes[0].set_xlabel("Reaction Time (s)", color=C1)
axes[0].set_ylabel("Probability Density", color=C1)
axes[0].set_title("Probability Regions", color=C1, fontweight='bold')
axes[0].legend(fontsize=9, facecolor='white', edgecolor=C1)
axes[0].tick_params(colors=C1)

bar_colours = [C2 if v < 0.5 else (C5 if v > 1.5 else C3) for v in pvals]
axes[1].barh([f'{p}th' for p in [5,10,25,50,75,90,95,99]], pvals,
             color=bar_colours, height=0.6, edgecolor='white')
axes[1].axvline(0.5, color=C2, linestyle='--', linewidth=2, label='0.5s threshold')
axes[1].axvline(1.5, color=C5, linestyle='--', linewidth=2, label='1.5s threshold')
axes[1].axvline(mu,  color=C1, linestyle='--', linewidth=2, label=f'mu = {mu}s')
for i, v in enumerate(pvals):
    axes[1].text(v + 0.01, i, f'{v:.3f}s', va='center', fontsize=9, color=C1, fontweight='bold')
axes[1].set_xlabel("Reaction Time (s)", color=C1)
axes[1].set_title("Key Percentile Values", color=C1, fontweight='bold')
axes[1].legend(fontsize=9, facecolor='white', edgecolor=C1)
axes[1].set_xlim(0, 1.7)
axes[1].tick_params(colors=C1)
plt.tight_layout()
plt.savefig('task3_probability.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
print("\n" + "=" * 60)
print("TASK 4: STATISTICS - ISA Speed Compliance")
print("=" * 60)

speeds = np.array([49, 52, 55, 48, 50, 53, 57, 51, 54, 49, 52, 56])
limit  = 50.0
n      = len(speeds)

mean_s    = np.mean(speeds)
median_s  = np.median(speeds)
std_s     = np.std(speeds, ddof=1)
sem_s     = std_s / np.sqrt(n)
exceeding = np.sum(speeds > limit)

print(f"\nSample: {speeds}")
print(f"\nDescriptive Statistics:")
print(f"  n              = {n}")
print(f"  Mean           = {mean_s:.2f} km/h")
print(f"  Median         = {median_s:.2f} km/h")
print(f"  Std deviation  = {std_s:.4f} km/h")
print(f"  Std error      = {sem_s:.4f} km/h")
print(f"  Min / Max      = {speeds.min()} / {speeds.max()} km/h")
print(f"  Over the limit = {exceeding} out of {n} ({exceeding/n*100:.1f}%)")


t_stat, p_value = stats.ttest_1samp(speeds, 50)
ci = stats.t.interval(0.95, df=n-1, loc=mean_s, scale=sem_s)
tc = stats.t.ppf(0.975, df=n-1)

print(f"\nOne-Sample t-Test  H0: mean = 50 km/h")
print(f"  t-statistic    = {t_stat:.4f}")
print(f"  p-value        = {p_value:.6f}")
print(f"  Critical val   = +/- {tc:.4f}  (alpha=0.05, df={n-1})")
print(f"  95% CI         = ({ci[0]:.2f}, {ci[1]:.2f}) km/h")
print(f"  Result         = {'REJECT H0 - significant speeding detected' if p_value < 0.05 else 'FAIL TO REJECT H0'}")

fig, axes = plt.subplots(1, 3, figsize=(14, 5))
fig.suptitle("Task 4: ISA Speed Compliance (Limit = 50 km/h)",
             fontsize=13, fontweight='bold', color=C1)

bar_cols = [C2 if s > limit else C3 for s in speeds]
axes[0].bar(range(1, n+1), speeds, color=bar_cols, edgecolor='white')
axes[0].axhline(limit,  color=C1, linestyle='--', linewidth=2.5, label='Limit 50')
axes[0].axhline(mean_s, color=C2, linestyle='-',  linewidth=2,   label=f'Mean {mean_s:.1f}')
axes[0].set_xlabel("Vehicle", color=C1)
axes[0].set_ylabel("Speed (km/h)", color=C1)
axes[0].set_title("Observed Speeds", color=C1, fontweight='bold')
axes[0].set_ylim(44, 62)
axes[0].legend(handles=[
    mpatches.Patch(color=C2, label='Exceeding'),
    mpatches.Patch(color=C3, label='Compliant'),
    plt.Line2D([0],[0], color=C1, linestyle='--', lw=2.5),
    plt.Line2D([0],[0], color=C2, lw=2)],
    labels=['Exceeding','Compliant','Limit 50','Mean 52.2'],
    fontsize=8, facecolor='white', edgecolor=C1)
axes[0].tick_params(colors=C1)

x_fit = np.linspace(44, 62, 200)
axes[1].hist(speeds, bins=6, color=C3, edgecolor='white', density=True, alpha=0.75, label='Observed')
axes[1].plot(x_fit, stats.norm.pdf(x_fit, mean_s, std_s), color=C1, linewidth=2.5, label=f'N({mean_s:.1f},{std_s:.1f})')
axes[1].axvline(limit,  color=C2, linestyle='--', linewidth=2.5, label='Limit 50')
axes[1].axvline(mean_s, color=C1, linewidth=1.5)
axes[1].set_xlabel("Speed (km/h)", color=C1)
axes[1].set_ylabel("Density", color=C1)
axes[1].set_title("Speed Distribution", color=C1, fontweight='bold')
axes[1].legend(fontsize=8, facecolor='white', edgecolor=C1)
axes[1].tick_params(colors=C1)

t_x   = np.linspace(-5, 5, 500)
t_pdf = stats.t.pdf(t_x, df=n-1)
axes[2].plot(t_x, t_pdf, color=C1, linewidth=2.5, label=f't-dist (df={n-1})')
axes[2].fill_between(t_x, t_pdf, where=(t_x >=  tc), alpha=0.5, color=C2, label='Rejection region')
axes[2].fill_between(t_x, t_pdf, where=(t_x <= -tc), alpha=0.5, color=C2)
axes[2].axvline(t_stat,  color=C4, linewidth=3,   label=f't = {t_stat:.3f}')
axes[2].axvline( tc, color=C1, linestyle='--', linewidth=1.5, label=f'Critical +/- {tc:.3f}')
axes[2].axvline(-tc, color=C1, linestyle='--', linewidth=1.5)
axes[2].set_xlabel("t value", color=C1)
axes[2].set_ylabel("Density", color=C1)
axes[2].set_title(f"t-Test (p = {p_value:.4f})", color=C1, fontweight='bold')
axes[2].legend(fontsize=8, facecolor='white', edgecolor=C1)
axes[2].tick_params(colors=C1)
plt.tight_layout()
plt.savefig('task4_statistics.png', dpi=150, bbox_inches='tight')
plt.show()
```
